# Рекомендация тарифов

В нашем распоряжении данные о поведении клиентов, которые уже перешли на эти тарифы. Нужно построить модель для задачи классификации, которая выберет подходящий тариф. Предобработка данных не понадобится — мы её уже сделали.

Построим модель с максимально большим значением *accuracy*.

## Откроем и изучим файл

```python
import pandas as pd
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/datasets/users_behavior.csv')
df.info()
df.head(10)
```

Все данные успешно выгружены. Все типы столбцов приведены к верному типу данных, также отсутвуют пропуски, что не может не радовать, так что предобрабокту опустим и сразу перейдем к разбивке выборок.

## Разобьем данные на выборки

Поскольку тестовой выборки нет, то будем делить наш разбивать наш датасет в соотношении 3:1:1, где 3 - тренировочная выборка, 1 - валидационная и 1 - тестовая. Для справедливости разбивки укажем параметр stratify.

```python
features = df.drop('is_ultra', axis=1)
target = df['is_ultra']

features_for_split, features_test, target_for_split, target_test = train_test_split(features, target, test_size=0.2,
                                                                                    random_state=12345, stratify=target)

features_train, features_valid, target_train, target_valid = train_test_split(features_for_split, target_for_split,
                                                                              test_size=0.25, random_state=12345, stratify=target_for_split)
```

```python
display(features_test)
display(target_test)
```

```python
display(features_train)
display(features_valid)
display(target_train)
display(target_valid)
```

```python
display(features_train.shape)
display(features_valid.shape)
display(features_test.shape)
display(target_train.shape)
display(target_valid.shape)
display(target_test.shape)
```

## Исследуем модели

Создадим словарь, чтобы записывать в него лучшие модели и их точности на валидационной выборке.

```python
dict_model = {}
```

Поскольку дерево сильно склонно к переобучению, то установим небольшую глубину 2-5.

```python
best_model_tree = None
best_result_tree = 0
arr_acc_train = list()
arr_acc_valid = list()

for depth in range(2, 4):
    model_tree = DecisionTreeClassifier(max_depth=depth, random_state=12345)
    model_tree.fit(features_train, target_train)
    arr_acc_train.append(model_tree.score(features_train, target_train))
    predictions = model_tree.predict(features_valid)
    result = accuracy_score(target_valid, predictions)
    arr_acc_valid.append(result)
    if result > best_result_tree:
        best_result_tree = result
        best_model_tree = model_tree
dict_model[best_model_tree] = best_result_tree

print("Лучшая точность:", best_result_tree)
```

```python
df_plot = pd.DataFrame({'valid_accuracy': arr_acc_valid, 'train_accuracy': arr_acc_train, 'depth': range(2, 4)})

df_plot.plot(x='train_accuracy', y='depth')
plt.ylabel('depth')
plt.legend(['accuracy'])
plt.show()
df_plot.plot(x='valid_accuracy', y='depth')
plt.ylabel('depth')
plt.legend(['accuracy'])
plt.show()
```

Точность 0.76 для одного дерева - впечатляет, посмотрим как другие проявят себя.

Также отобразим нашу модель в виде понятного графика.

```python
plt.figure(figsize=(30, 30))
print(features.columns)
_ = tree.plot_tree(best_model_tree,
                   feature_names=features.columns,
                   class_names=['0', '1'],
                   filled=True)
plt.show()
```

Тут уже лес, но он также склонен к переобучению, хоть и не так сильно как отдельное дерево, поэтому укажем глубину здесь побольше: 3-9, и количество деревьев 3-7.

```python
best_model_forest = None
best_result_forest = 0
arr_acc_train = list()
arr_acc_valid = list()
for trees in range(3, 8):
    for depth in range(3, 10):
        model_forest = RandomForestClassifier(random_state=12345, n_estimators=trees, max_depth=depth)
        model_forest.fit(features_train, target_train)
        arr_acc_train.append(model_forest.score(features_train, target_train))
        predictions = model_forest.predict(features_valid)
        result = accuracy_score(target_valid, predictions)
        arr_acc_valid.append(result)
        if result > best_result_tree:
            best_result_forest = result
            best_model_forest = model_forest
dict_model[best_model_forest] = best_result_forest
```

```python
df_plot = pd.DataFrame({'valid_accuracy': arr_acc_valid, 'train_accuracy': arr_acc_train})

plt.plot(df_plot['valid_accuracy']);
plt.legend(['valid_accuracy'])
plt.show()


plt.plot(df_plot['train_accuracy']);
plt.legend(['train_accuracy'])
plt.show()
```

Тут можно поиграться с количеством итераций, однако после 100 точность не меняется.

```python
best_model_logic = None
best_result_logic = 0

model_logic = LogisticRegression(random_state=12345, solver='lbfgs', max_iter=100)
model_logic.fit(features_train, target_train)
arr_acc_train.append(model_logic.score(features_train, target_train))
result = model_logic.score(features_valid, target_valid)
arr_acc_valid.append(result)
dict_model[model_logic] = result
```

```python
dict_model
```

Таким образом, наибольшую точность на этапе валидационной выборки показывает случайны лес, затем дерево, а только потом логическая регрессия. Ожидание были другие - 3 и 2 место поменять местами.

## Проверим модель на тестовой выборке

Хоть есть и фаворит - Случайный лес, все равно было бы неплохо пройтись по всем моделям, вдруг что-то измениться.

```python
best_model = None
best_result = 0
for models, results in dict_model.items():
    if models.score(features_test, target_test) > best_result:
        best_result = models.score(features_test, target_test)
        best_model = models
        print(models, models.score(features_test, target_test))

print("Лучшая модель:", best_model)
print("Точность этой модели:", best_result)
```

Точность 0.81 - довольно неплохо при таком малом урегулировании параметров моделей.

Теперь, после того, как наши модели обучены, все проверены на тестовой выборке, а валидационная выборка сделала свою работу, объединим тренировочную и валидационную выборки, и уже обучим модели с нашими подобранными гиперпараметрами на этой увеличенной выборке.

```python
dict_model
```

Создадим новый словарь куда будем записывать новые обученные модели.

```python
new_dict_model = {}
```

```python
features_full_train = pd.concat([features_train, features_valid])
target_full_train = pd.concat([target_train, target_valid])
```

```python
model_tree = DecisionTreeClassifier(max_depth=3, random_state=12345)
model_tree.fit(features_full_train, target_full_train)
new_dict_model[model_tree] = model_tree.score(features_test, target_test)
print(model_tree.score(features_test, target_test))
```

```python
model_forest = RandomForestClassifier(max_depth=9, random_state=12345, n_estimators=7)
model_forest.fit(features_full_train, target_full_train)
new_dict_model[model_forest] = model_forest.score(features_test, target_test)

print(model_forest.score(features_test, target_test))
```

```python
model_logistic = LogisticRegression(random_state=12345, solver='lbfgs', max_iter=100)
model_logistic.fit(features_full_train, target_full_train)
new_dict_model[model_logistic] = model_logistic.score(features_test, target_test)

print(model_logistic.score(features_test, target_test))
```

```python
new_dict_model
```

```python
print("Точность старой модели Дерева на тестовой выборке составляла: 0.7993779160186625")
print("Точность старой модели Леса на тестовой выборке составляла: 0.8118195956454122")
print("Точность старой модели Регрессии на тестовой выборке составляла: 0.7371695178849145")
```

Результаты изменились: для всех моделей кроме Леса точность упала, а вот у Леса совсем немного выросла.

Проверим модели на адекватность

```python
print(len(df))
print(len(target_test))

dummy_prediction = pd.Series([0] * len(target_test))
dummy_accuracy = accuracy_score(target_test, dummy_prediction)
dummy_accuracy
```

После проверки на адекватность "глупая" модель выдала точность 0.69! Эта точность конечно ниже нашего фаворита - 0.81, однако недалеко ушло, есть над чем поработать и постараться улучшить нашу модель.
